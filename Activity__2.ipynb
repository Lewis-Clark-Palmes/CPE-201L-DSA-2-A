```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr):
    print("reversing all elements")
    reversed_arr = arr[::-1]
    print(reversed_arr)

def insert(arr):
    print("inserting element")
    arr.insert(1, 3)
    print(arr)

def delete(arr):
    print("deleting element")
    del arr[2]
    print(arr)

def modify(arr):
    print("modifying element")
    arr[2] = 300
    print(arr)

def sort(arr):
    print("sorting element")
    arr.sort()
    print(arr)

def append(arr):
    print("appending element")
    arr.append(60)
    print(arr)

def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Exit")
        print("3. Reverse array")
        print("4. Insert element")
        print("5. Delete element")
        print("6. Modify element")
        print("7. Sort array")
        print("8. Append element")
        choice = input("Enter your choice (1-8): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '3':
            reverse(arr)
        if choice == '4':
            insert(arr)
        elif choice == '5':
            delete(arr)
        elif choice == '6':
            modify(arr)
        elif choice == '7':
            sort(arr)
        elif choice == '8':
            append(arr)
        if choice == '2':
            print("Exiting the program.")
            break
        else:

          print("Invalid choice. Please enter a number between 1 to 8.")


if __name__ == "__main__":
    main()
```
